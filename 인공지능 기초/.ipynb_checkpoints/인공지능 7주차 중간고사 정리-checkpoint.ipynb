# 표안의 그래프의 이름, 여러 그래프 그리기.

```python
a = int(input('a를 입력하세요')) # ''안에 내가 원하는 문구를 넣어서 문구를 띄울수 있음. 

x = np.arange(-10, 10, 0.02)  # arange는 a range 말과 같다. 
y = 1/(1 + np.exp(-a*x)) 
plt.plot(x, y)
plt.grid(alpha=0.3)
```

```python
x = np.arange(-10, 10, 0.02)  # arange는 a range 말과 같다. 

for a in range(1,5):  
    y = 1/(1 + np.exp(-a*x))
    plt.plot(x, y, label=a)
    
plt.grid(alpha=0.3)
plt.legend()

# label=a 적고 맨 아래에 plt.legend() 했을경우 오른쪽 위에 선이 어떤 선인지 적어준다. 
```

# sin cos tan 그래프 그리기

```python
# sin, cos 그래프 그리기

x = np.linspace(0, 2*np.pi, 500)
y = np.sin(x)
plt.plot(x,y, label='y=sin(x)') # label = '00000' 이런식으로 적을경우 왼쪽 밑에 글씨가 뜬다.

y = np.cos(x)
plt.plot(x, y, label='x=cos(x)')
plt.legend(loc='upper center'); # plt.legend 를 하면 label 지정한 글씨들이 뜨게 만든다.
```

```python
# sin, cos 그래프 그리기 심화
x = np.linspace(-4*np.pi, 4*np.pi, 500) # -4*np.pi, 4*np.pi -> 범위를 지정해주는 것이다. 
y = np.sin(x)
plt.plot(x,y, label='y=sin(x)') # label = '00000' 이런식으로 적을경우 왼쪽 밑에 글씨가 뜬다.

y = np.cos(x)
plt.plot(x, y, label='x=cos(x)')
plt.ylim(-2, 2) # y축의 범위를 지정해줘서 더 넓게 보거나, 좁게 보거나 할수 있다.
plt.grid(alpha = 0.3) # 표 안에 격자를 넣는것이다.
plt.legend(); # plt.legend 를 하면 label 지정한 글씨들이 뜨게 만든다.
```

```python
# tan 그리기
x = np.linspace(0, 2*np.pi, 500)
y = np.tan(x)
plt.xlim(-4,4)
plt.ylim(-8,8)
plt.plot(x, y);
```

# 점과 점사이의 거리 구하기

```python
def e_distance(a1, a2, b1, b2): # 2차원에서의 거리를 구할땐 이런식으로 구한다.
    e_distance=np.sqrt(np.square(b1-a1) + np.square(b2-a2))
    return e_distance

e_distance(0, 0, 4, 3)
```

```python
from scipy.spatial import distance # scipy = 과학용을 계산하는 파이썬으로 생각하면 된다.

p1 = (0, 0)
p2 = (4, 3)
e_distance = distance.euclidean(p1, p2)
e_distance
# 바로 위의 함수보다 scipy를 사용하면 좀더 간단하게 구할수 있다  
```

# 등차수열 코딩으로 해보기

```python
# 등차수열 코딩으로 하기

n = 10 # n번째 
a = 4  # 초항
d = 5  # 5씩 커지는
 
sum = 0
for n in range(1,n+1):
    sum = sum + (a + (n-1)*d)
sum
```

# 수열

```python
# 수열 더하기
total = 0

for i in range(1, 6):
    total += 2*i-1
print(total)
```

```python
# 수열의 곱하기
total = 1

for i in range(1, 6):
    total *= 2*i-1
print(total)
```

```python
# 다른 방법으로 구하는 방법 (reduce)
from functools import reduce
a = list(range(1, 6))
reduce(lambda acc, cur:acc*cur, a, 1)
```

```python
# 다른 방법으로 구하는 방법 (reduce 넘파이로 하기)
from functools import reduce
a = np.arange(1 ,6)
reduce(lambda acc, cur:acc*(2*cur-1), a, 1)
```

# 교집합, 합집합 등등등 

```python
A = {1, 2, 3, 4, 5}
B = {4, 5, 6 ,7}
```

```python
A&B # 교집합
A|B # 합집합
A-B # 차집합
A^B # 대칭차집합
```

```python
A.intersection(B) # 교집합
A.union(B) # 합집합
A.difference(B) # 차집합 
A.symmetric_difference(B) # 대칭차집합
```

```python
np.intersect1d(A, B) # 넘파이 같은 경우에는 집합 기호를 사용하면 안됨. (배열 데이터로 사용을 해줘야 가능하다.)
```

```python
A = np.array([1, 2, 3, 4, 5]) 
B = np.array([4, 5, 6 ,7])
```

```python
np.intersect1d(A, B)
```

# 도함수와 그래프 그리기(?)

```python
# y = x³+3x²+3x+1 의 도함수와 그래프를 그려본다고 하심.
# 도함수 = y'=3x²+6x+3 <- 이걸 그림으로 일단 그려본다고 하심
# 일단 x데이터가 있어야하니까 x값을 가져와야함 
```

```python
x = np.arange(-30, 30, 0.1)
y = x**3 + 3*x**2 + 3*x + 1
plt.plot(x, y, label = 'f(x)');
#plt.grid(alpha=0.3, color = 'y') # plt.grid -> 그래프에 격자를 넣어주는것, (alpha = 0.3 - 투명도를 0.3 만큼 한다.)

y1 = 2*x**2 + 6*x
plt.plot(x, y1, label = "f'(x)");
#plt.xlim(-30, 30) # 그래프의 범위를 바꿔주는 명령
plt.ylim(-4000, 4000) # 범위에 따라서 겹치는 부분이 보였다 안보였다가 한다.... 
plt.legend() # 그래프 안에 선들을 구분할수 있도록 박스안에 이름을 넣어준다. 
plt.grid(alpha=0.3)
```
