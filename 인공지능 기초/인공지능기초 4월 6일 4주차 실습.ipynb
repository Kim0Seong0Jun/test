```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = int(input('a를 입력하세요'))

x = np.arange(-10, 10, 0.02)  # arange는 a range 말과 같다. 
y = 1/(1 + np.exp(-a*x))
plt.plot(x, y)
plt.grid(alpha=0.3)
```

```python
x = np.arange(-10, 10, 0.02)  # arange는 a range 말과 같다. 

for a in range(1,5):  
    y = 1/(1 + np.exp(-a*x))
    plt.plot(x, y, label=a)
    
plt.grid(alpha=0.3)
plt.legend()

# label=a 적고 맨 아래에 plt.legend() 했을경우 오른쪽 위에 선이 어떤 선인지 적어준다. 
```
