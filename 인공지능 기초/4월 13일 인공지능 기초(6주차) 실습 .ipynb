```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 
x = np.linspace(0, 2*np.pi, 500)
y = np.sin(x)
plt.plot(x,y, label='y=sin(x)') # label = '00000' 이런식으로 적을경우 왼쪽 밑에 글씨가 뜬다.

y = np.cos(x)
plt.plot(x, y, label='x=cos(x)')
plt.legend(loc='upper center'); # plt.legend 를 하면 label 지정한 글씨들이 뜨게 만든다.
```

```python
# 
x = np.linspace(-4*np.pi, 4*np.pi, 500) # -4*np.pi, 4*np.pi -> 범위를 지정해주는 것이다. 
y = np.sin(x)
plt.plot(x,y, label='y=sin(x)') # label = '00000' 이런식으로 적을경우 왼쪽 밑에 글씨가 뜬다.

y = np.cos(x)
plt.plot(x, y, label='x=cos(x)')
plt.ylim(-2, 2) # y축의 범위를 지정해줘서 더 넓게 보거나, 좁게 보거나 할수 있다.
plt.grid(alpha = 0.3) # 표 안에 격자를 넣는것이다.
plt.legend(); # plt.legend 를 하면 label 지정한 글씨들이 뜨게 만든다.
```

```python
def e_distance(a1, a2, b1, b2): # 2차원에서의 거리를 구할땐 이런식으로 구한다.
    e_distance=np.sqrt(np.square(b1-a1) + np.square(b2-a2))
    return e_distance

e_distance(0, 0, 4, 3)
# 하지만 3차원에서의 거리를 구할때는 위의 함수로 구하기엔 복잡해진다.
# 모듈이 모이면은 라이브러리가 되는것이다. 크기로 따지게 된다면 라이브러리가 더 큰것이다.
```

```python
from scipy.spatial import distance # scipy = 과학용을 계산하는 파이썬으로 생각하면 된다.

p1 = (0, 0)
p2 = (4, 3)
e_distance = distance.euclidean(p1, p2)
e_distance
# 바로 위의 함수보다 scipy를 사용하면 좀더 간단하게 구할수 있다    
```
