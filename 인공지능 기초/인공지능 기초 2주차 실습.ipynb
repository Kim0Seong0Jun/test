```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [3, 5, 7, 9]
#plt 매플럿안에 있는 함수 plot를 사용한다는 말이다. 
plt.plot(x, y);
# plt.show()
```

```python

y=[]
x=list(range(1, 5))
for i in range(1, 5):
    y.append(2*i+1)
plt.plot(x, y)
```

```python
y=[]
x=list(range(1, 5)) 
for i in x:
    y.append(2*i+1)
plt.plot(x, y);
```

```python
import numpy as np
#파이썬에서는 레인지를 사용했고, 넘파인에서는 어레인지를 사용한다, 범위가 -1 부터 1까지 간격은 0.01로 한다는 말이다. 

X = np.arange(-1, 1, 0.01) 
Y = 2*X + 1
plt.plot(X, Y, 'k--'); 



#이렇게 계산할때는 넘파인이 계산하기 편하고 빠르다. 
```

```python
X = np.linspace(-2, 2, 1001) #여기서 1001는 갯수를 말한다.
Y = 2*X + 1
plt.plot(X, Y, 'k--');
```
